# Try BKT

The model will be fitted on data from all students to better learn the underlying learning process. However, the prediction will be unique for each students based on their unique sequence of responses.

## Load Data

```python
import pandas as pd

# Read CSV ignore index
data = pd.read_csv("../data/sample_sep.csv", index_col=0)
```

```python
data.head()
```

```python
# Check the response from one specific student
data[data["responses_lo_2.3.student_id"] == "bc714b09-c2a9-414c-8f22-14c1f63eb297"]
```

```python
# Run some transformations to get the data into the correct format

# Drop nan values in the responses_lo_2.3.points_earned column
data = data.dropna(subset=["responses_lo_2.3.points_earned"])

# Ensure the dt_submitted column is in datetime format
data["responses_lo_2.3.dt_submitted"] = pd.to_datetime(
    data["responses_lo_2.3.dt_submitted"]
)

# Sort the data by student and submission time to create order_id
data = data.sort_values(
    by=["responses_lo_2.3.student_id", "responses_lo_2.3.dt_submitted"]
)

# Create order_id (sequence number of observations per student)
data["order_id"] = data.groupby("responses_lo_2.3.student_id").cumcount() + 1

# Rename columns to match pyBKT expected format
bkt_data = data.rename(
    columns={
        "responses_lo_2.3.student_id": "user_id",
        "responses_lo_2.3.points_earned": "correct",
    }
)

# Assign a skill name (you can change this if you're tracking multiple skills)
bkt_data["skill_name"] = 0  # If you're tracking only one skill, this can be constant

# Drop unnecessary columns
bkt_data = bkt_data[["user_id", "order_id", "correct", "skill_name"]]
```

```python
from pyBKT.models import Model

model = Model(seed=42, num_fits=1)
model.fit(data=bkt_data)
```

```python
# Params being used
# TODO: Understand how these are learned from the data
model.coef_
```

```python
# Generate predictions for each attempt
predictions = model.predict(data=bkt_data)
```

```python
predictions
```

```python
# Get the final predictions after all attempts for each student
final_predictions = predictions.groupby("user_id").last()[
    ["correct_predictions", "state_predictions"]
]

# Rename the aggregated columns
final_predictions = final_predictions.rename(
    columns={
        "state_predictions": "final_state_predictions",
        "correct_predictions": "final_correct_predictions",
    }
)
```

```python
final_predictions
```

```python
# Store the result in a CSV file
final_predictions.to_csv("sample_sep_final_predictions.csv")
```
